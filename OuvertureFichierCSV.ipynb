```python
import csv 
def statsCSV(fichier):
    
    file = open(fichier,"r")
    donnees = csv.reader(file)
    ligne = list(donnees)
    nb_ligne= len(ligne)
    nb_colonne=len(ligne[0])
    return  nb_ligne, nb_colonne
file.close()
print(statsCSV("BDtest.csv"))
f= open (r"BDtest.csv") 
myReader = csv.reader(f)  
for row in myReader:     
    print(row)
   
```
